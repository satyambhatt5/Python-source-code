```python
# Solution:
import pandas as pd
import numpy as np
import scipy.stats as stats
import math
spend = pd.read_csv("C:\\Users\\aagarwal\\Desktop\\DataSetsR\\Stat_Test_Input.csv") 
```

```python
# Q1 - find if the spending behavoir of males is different then females ? 
           # Independent Sample T TEST | Akshay

# Null Hypothesis: Equality | Mean of S1 = Mean of S2
               # The spending of male and female is same
    
Males_Spending = spend.Spending_Before_Marriage[spend.Gender==0]  
Females_Spending = spend.Spending_Before_Marriage[spend.Gender==1] 
```

```python
spend.columns

Females_Spending.shape
```

```python
print(Females_Spending.mean())# 35.195945945945944
print(Males_Spending.mean()) # 35.562913907284766


print(spend.Spending_Before_Marriage.mean())
print(spend.Spending_after_Marriage.mean())
```

```python
# Independent Sample T TEst:
stats.ttest_ind(a=Males_Spending,
                b=Females_Spending)

# pvalue=0.03892731792383138 [ P is lower then 0.05 then Null will GO - Reject]
# Conclusion : Spending of Female is different Then Males
```

#### Q2 : # Q2 - find if there is a change in the spedning before marriage and afer marriage ? Hint : Paired T TESt 

```python
# NULL - Mean of Spend Before Marriage = mean of Spend after marriage

stats.ttest_rel(a=spend.Spending_Before_Marriage,
               b=spend.Spending_after_Marriage)

# pvalue=1.5036002343854017e-07 | P is low (lower then 0.05) | Reject the N0
# Conclusion | The spending before and after marriage is different


```
